```python
from esridumpgdf import Site, Service, Layer
```

### Site methods
Use the `Site` class to introspect an ArcGIS Server instance

```python
site = Site('https://sampleserver6.arcgisonline.com/arcgis/rest/services')
site
```

```python
services = site.services()
services
```

```python
layers = site.layers()
layers
```

### Service methods
Use the `Service` class to introspect and export an ArcGIS Server service

```python
service = Service(url=services.iloc[1].url)
service
```

```python
service.layers()
```

#### Export a list of geodataframes mapping to the service's layers

```python
gdfs = service.to_gdfs()
gdfs
```

#### Export the service to geopackage

```python
service.to_gpkg(filename='demo.gpkg')
```
